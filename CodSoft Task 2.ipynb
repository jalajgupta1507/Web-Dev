```python
import tkinter as tk
from tkinter import messagebox

def calculate():
    try:
        num1 = float(entry_num1.get())
        num2 = float(entry_num2.get())
        operation = operation_choice.get()

        if operation == "Addition":
            result = num1 + num2
        elif operation == "Subtraction":
            result = num1 - num2
        elif operation == "Multiplication":
            result = num1 * num2
        elif operation == "Division":
            if num2 != 0:
                result = num1 / num2
            else:
                messagebox.showerror("Error", "Cannot divide by zero!")
                return
        else:
            messagebox.showerror("Error", "Invalid operation!")
            return

        result_label.configure(text="Result: " + str(result))
    except ValueError:
        messagebox.showerror("Error", "Invalid input!")

# Create the main window
window = tk.Tk()
window.title("Calculator")
window.configure(bg="#f2f2f2")

# Create labels and entry fields for number 1
label_num1 = tk.Label(window, text="Number 1:", font=("Helvetica", 14), bg="#f2f2f2")
label_num1.grid(row=0, column=0, padx=10, pady=10)
entry_num1 = tk.Entry(window, font=("Helvetica", 12))
entry_num1.grid(row=0, column=1, padx=10, pady=10)

# Create labels and entry fields for number 2
label_num2 = tk.Label(window, text="Number 2:", font=("Helvetica", 14), bg="#f2f2f2")
label_num2.grid(row=1, column=0, padx=10, pady=10)
entry_num2 = tk.Entry(window, font=("Helvetica", 12))
entry_num2.grid(row=1, column=1, padx=10, pady=10)

# Create label and operation choice
label_operation = tk.Label(window, text="Operation:", font=("Helvetica", 14), bg="#f2f2f2")
label_operation.grid(row=2, column=0, padx=10, pady=10)
operation_choice = tk.StringVar()
operation_choice.set("Addition")
choices = ["Addition", "Subtraction", "Multiplication", "Division"]
operation_menu = tk.OptionMenu(window, operation_choice, *choices)
operation_menu.config(font=("Helvetica", 12))
operation_menu.grid(row=2, column=1, padx=10, pady=10)

# Create the calculate button
calculate_button = tk.Button(window, text="Calculate", font=("Helvetica", 14), command=calculate)
calculate_button.grid(row=3, column=0, columnspan=2, padx=10, pady=20)

# Create the result label
result_label = tk.Label(window, text="Result: ", font=("Helvetica", 16), bg="#f2f2f2")
result_label.grid(row=4, column=0, columnspan=2)

# Run the main event loop
window.mainloop()
```
