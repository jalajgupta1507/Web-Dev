```python
import tkinter as tk
from tkinter import messagebox
import random
import string

def generate_password():
    try:
        length = int(entry_length.get())
        if length <= 0:
            messagebox.showerror("Error", "Invalid password length!")
            return

        username = entry_username.get().strip()

        if username:
            password = generate_random_password(length)
            password_label.configure(text="Generated Password: " + password)
        else:
            messagebox.showerror("Error", "Please enter a username!")

    except ValueError:
        messagebox.showerror("Error", "Invalid input!")

def generate_random_password(length):
    chars = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(random.choices(chars, k=length))
    return password

def accept_password():
    accepted_password = password_label.cget("text").split(": ")[1]
    messagebox.showinfo("Accepted", "Accepted Password: " + accepted_password)

def reset_password():
    password_label.configure(text="Generated Password: ")

# Create the main window
window = tk.Tk()
window.title("Password Generator")
window.configure(bg="#f2f2f2")

# Create label and entry field for username
label_username = tk.Label(window, text="Username:", font=("Helvetica", 14), bg="#f2f2f2")
label_username.pack(pady=10)
entry_username = tk.Entry(window, font=("Helvetica", 12))
entry_username.pack()

# Create label and entry field for password length
label_length = tk.Label(window, text="Password Length:", font=("Helvetica", 14), bg="#f2f2f2")
label_length.pack(pady=10)
entry_length = tk.Entry(window, font=("Helvetica", 12))
entry_length.pack()

# Create the generate button
generate_button = tk.Button(window, text="Generate Password", font=("Helvetica", 14), command=generate_password)
generate_button.pack(pady=20)

# Create the password label
password_label = tk.Label(window, text="Generated Password: ", font=("Helvetica", 16), bg="#f2f2f2")
password_label.pack()

# Create the accept and reset buttons
accept_button = tk.Button(window, text="Accept Password", font=("Helvetica", 12), command=accept_password)
accept_button.pack(pady=10)

reset_button = tk.Button(window, text="Reset Password", font=("Helvetica", 12), command=reset_password)
reset_button.pack(pady=5)

# Run the main event loop
window.mainloop()
```
