```python
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk


class TodoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, task):
        self.tasks.append(task)
        messagebox.showinfo("Success", "Task added successfully!")

    def delete_task(self, task):
        if task in self.tasks:
            self.tasks.remove(task)
            messagebox.showinfo("Success", "Task deleted successfully!")
        else:
            messagebox.showerror("Error", "Task not found in the list.")

    def update_task(self, task, new_task):
        if task in self.tasks:
            index = self.tasks.index(task)
            self.tasks[index] = new_task
            messagebox.showinfo("Success", "Task updated successfully!")
        else:
            messagebox.showerror("Error", "Task not found in the list.")

    def display_tasks(self):
        if self.tasks:
            task_list = "\n".join(self.tasks)
            messagebox.showinfo("Task List", task_list)
        else:
            messagebox.showinfo("Task List", "No tasks found in the list.")


def add_task():
    task = task_entry.get()
    if task:
        todo_list.add_task(task)
        task_entry.delete(0, tk.END)
    else:
        messagebox.showerror("Error", "Please enter a task.")

def delete_task():
    task = task_entry.get()
    if task:
        todo_list.delete_task(task)
        task_entry.delete(0, tk.END)
    else:
        messagebox.showerror("Error", "Please enter a task.")

def update_task():
    task = task_entry.get()
    new_task = new_task_entry.get()
    if task and new_task:
        todo_list.update_task(task, new_task)
        task_entry.delete(0, tk.END)
        new_task_entry.delete(0, tk.END)
    else:
        messagebox.showerror("Error", "Please enter both the task and new task.")

def display_tasks():
    todo_list.display_tasks()

# Create the main window
window = tk.Tk()
window.title("To-Do List")
window.configure(bg="#f2f2f2")

# Create a TodoList object
todo_list = TodoList()

# Create the task entry label and entry field
task_label = tk.Label(window, text="Task:", font=("Helvetica", 14), bg="#f2f2f2")
task_label.pack()
task_entry = tk.Entry(window, font=("Helvetica", 12))
task_entry.pack()

# Create the new task entry label and entry field
new_task_label = tk.Label(window, text="New Task:", font=("Helvetica", 14), bg="#f2f2f2")
new_task_label.pack()
new_task_entry = tk.Entry(window, font=("Helvetica", 12))
new_task_entry.pack()

# Create buttons for adding, deleting, updating, and displaying tasks
add_button = ttk.Button(window, text="Add Task", command=add_task)
add_button.pack(pady=5)
delete_button = ttk.Button(window, text="Delete Task", command=delete_task)
delete_button.pack(pady=5)
update_button = ttk.Button(window, text="Update Task", command=update_task)
update_button.pack(pady=5)
display_button = ttk.Button(window, text="Display Tasks", command=display_tasks)
display_button.pack(pady=5)

# Run the main event loop
window.mainloop()
```
